*Cleaning the data

```python
#Check for null values that may be present in dataset

exams.isnull().sum()
```

```python
#removing columns that are not needed

exams = exams.drop(['race/ethnicity', 'parental level of education'], axis = 1)
exams.head()
```

```python
#Correlation analysis between exam scores

correlation = exams.corr()
correlation
```

```python
#Plotting correlation analysis

sb.heatmap(correlation, xticklabels = correlation.columns, yticklabels = correlation.columns, annot = True);
```

```python
#Using the seaborn pairplot to visualize the relationship between all variables in the dataset

sb.pairplot(exams);
```

```python
#Using scatterplot to show gender performance of students in Math and reading exams

sb.relplot(data = exams, x = 'math score', y = 'reading score', hue = 'gender');
```

```python
#Using a histogram to show the score variations of students in Math exam

sb.distplot(exams['math score']);
```

```python
#Using a boxplot to determine the prevailing scores of students in the writing exam

sb.catplot(data = exams, x = 'writing score', kind = 'box');
```
